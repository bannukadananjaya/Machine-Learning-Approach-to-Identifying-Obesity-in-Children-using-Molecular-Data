```python
non_zero_values = df[df.filter(like='Detection Pval').ne(0).any(axis=1)]

# Check if there are any rows with non-zero values
if not non_zero_values.empty:
    print("There are rows with non-zero Detection Pval values.")
else:
    print("All Detection Pval values are 0.")
```
